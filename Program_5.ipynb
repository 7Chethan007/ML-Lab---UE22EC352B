# Linear Regression with Gradient Descent
# Making the imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (12.0, 9.0)

# Preprocessing Input data
data = pd.read_csv('data.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]

plt.scatter(X, Y)
plt.show()

# Building the model
w1_old = 0.1
w0_old = 0.1
w1_new = 0
w0_new = 0
alpha = 0.0001  # Learning Rate
epochs = 1000   # Iterations
n = float(len(X))

# Performing Gradient Descent
while abs(w1_new - w1_old) + abs(w0_new - w0_old) > 0.01:
    w1_old = w1_new
    w0_old = w0_new
    Y_pred = w1_old * X + w0_old  # Current predicted Y

    D_w1 = (-2/n) * sum(X * (Y - Y_pred))  # Derivative w.r.t w1
    D_w0 = (-2/n) * sum(Y - Y_pred)        # Derivative w.r.t w0

    w1_new = w1_old - alpha * D_w1  # Update w1
    w0_new = w0_old - alpha * D_w0  # Update w0

    print(w1_new, w0_new)

# Making predictions
Y_pred = w1_new * X + w0_new
plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_regression_lines():
    X = np.linspace(0, 10, 100)

    # Different slope and intercept combinations
    scenarios = [
        (1, 0),   # Moderate positive slope, zero intercept
        (2, 2),   # Steeper positive slope, positive intercept
        (0.5, -1),# Gentle positive slope, negative intercept
        (-1, 3)   # Negative slope, positive intercept
    ]

    plt.figure(figsize=(12, 8))

    for i, (m, c) in enumerate(scenarios, 1):
        Y = m * X + c
        plt.subplot(2, 2, i)
        plt.scatter(X, Y)
        plt.plot(X, Y, 'r-')
        plt.title(f'm = {m}, c = {c}')
        plt.xlabel('X')
        plt.ylabel('Y')

    plt.tight_layout()
    plt.show()

plot_regression_lines()
```
